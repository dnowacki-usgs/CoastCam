```python
# csherwood@usgs.gov, 2026-07-17, generated with Sonnet 4.6

import pandas as pd
import matplotlib.pyplot as plt

fname = "F:/crs/proj/marconi_imagery/tide_model/tidal_data_marconi_modelled_20260715-20260831.csv"

df = pd.read_csv(fname, parse_dates=["time"])
df = df.set_index("time")

models = ["EOT20_heightm", "GOT5.5_heightm", "GOT5.6_heightm", "FES2022_heightm"]

fig, ax = plt.subplots(figsize=(10, 4))
for m in models:
    ax.plot(df.index, df[m], label=m.replace("_heightm", ""))

ax.set_xlabel("Time")
ax.set_ylabel("Height (m)")
ax.legend()
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
for m in models:
    ax.plot(df.index, df[m]-df["FES2022_heightm"], label=m.replace("_heightm", ""))

ax.set_xlabel("Time")
ax.set_ylabel("Height (m)")
ax.legend()
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

```python
others = [m for m in models if m != "FES2022_heightm"]

fig, ax = plt.subplots(figsize=(8, 5))
for m in others:
    diff = df[m] - df["FES2022_heightm"]
    ax.hist(diff, bins=30, alpha=0.5, label=m.replace("_heightm", "") + " - FES2022")

ax.set_xlabel("Height difference (m)")
ax.set_ylabel("Count")
ax.legend()
plt.tight_layout()
plt.show()
```
